```python
%%sql

CREATE VIEW view_revenue_analysis AS
SELECT o.id, o.user_id,o.paid_status, o.is_delivered, round(o.total_price*0.011, 2) AS total_price_usd,
o.ordered_item AS product, 
b.state, 1 AS count_of_product,
    CASE
        WHEN paid_status = 1 and is_delivered = 1 THEN 1
        ELSE 0
    END AS count_paid_delivered,
    CASE
        WHEN paid_status = 1 and is_delivered = 1 THEN round(o.total_price*0.011, 2)
        ELSE 0
    END AS revenue_paid_delivered,
   CASE
        WHEN paid_status = 1 and is_delivered = 0 THEN 1
        ELSE 0
    END AS count_paid_undelivered,
    CASE
        WHEN paid_status = 1 and is_delivered = 0 THEN round(o.total_price*0.011, 2)
        ELSE 0
    END AS revenue_paid_undelivered,
    CASE
        WHEN paid_status = 0 and is_delivered = 0 THEN 1
        ELSE 0
    END AS count_unpaid_undelivered,
    CASE
        WHEN paid_status = 0 and is_delivered = 0 THEN round(o.total_price*0.011, 2)
        ELSE 0
    END AS revenue_unpaid_undelivered,
    CASE
        WHEN paid_status = 0 and is_delivered = 1 THEN round(o.total_price*0.011, 2)
        ELSE 0
    END AS count_unpaid_delivered,
    CASE
        WHEN paid_status = 0 and is_delivered = 1 THEN round(o.total_price*0.011, 2)
        ELSE 0
    END AS revenue_unpaid_delivered
FROM account_ordermodel o
LEFT JOIN view_account_billingaddress b ON o.user_id = b.user_id
WHERE state <> 10;
```

```python
%%sql

SELECT *
FROM view_revenue_analysis
LIMIT 2;
```

Q2.  What is the distribution of the amount of products ORDERED? How much revenue would each product have generated IF the order was completed?

```python
%%sql

CREATE VIEW view_q2_distribution_of_products AS
SELECT product, count(*) as units_ordered, round(sum(total_price_usd), 2) as revenue
FROM view_revenue_analysis
Group By product
Order by units_ordered DESC;
```

```python
%%sql

SELECT *
FROM view_q2_distribution_of_products
LIMIT 10;
```

Q3A.  Which states ordered the most products? How does the revenue generated compare to the amount of products sold?

```python
%%sql

CREATE VIEW view_q3a_top_10_state AS
SELECT state, count(*) as units_ordered, round(sum(total_price_usd), 2) as revenue, round(sum(total_price_usd)/count(*), 2) as revenue_per_unit_ordered
FROM view_revenue_analysis
Group By state
ORDER BY units_ordered DESC
LIMIT 10;
```

```python
%%sql

SELECT *
FROM view_q3a_top_10_state;
```

Q3B.  Which states ordered the least products? How does the revenue generated compare to the amount of products sold? 

```python
%%sql

CREATE VIEW view_q3b_bottom_10_states AS
SELECT state, count(*) as units_ordered, round(sum(total_price_usd), 2) as revenue, round(sum(total_price_usd)/count(*), 2) as revenue_per_unit_ordered
FROM view_revenue_analysis
Group By state
ORDER BY units_ordered ASC
LIMIT 10;
```

```python
%%sql

SELECT *
FROM view_q3b_bottom_10_states;
```

```python
Q3C.  What were the top 5 products sold in each of those states? (Answer should include all states)
```

```python
%%sql

CREATE VIEW view_q3c_top_5_product_per_state AS
WITH RankedProducts AS (
    SELECT
        state,
        product,
        ROUND(SUM(total_price_usd)) AS revenue,
        COUNT(*) AS units_ordered,
        RANK() OVER (PARTITION BY state ORDER BY COUNT(*) DESC) AS product_rank
    FROM
        view_revenue_analysis
    GROUP BY
        state, product
)

SELECT
    state,
    product,
    revenue,
    units_ordered,
    product_rank
FROM
    RankedProducts
WHERE
    product_rank <= 5;
```

```python
%%sql

SELECT * 
FROM view_q3c_top_5_product_per_state
```

Q4. There are orders that were placed that weren't paid for. Those shouldn't have been shipped out. How much money did this company lose by delivering products that were not paid for? How does this compare to how much revenue would have been generated?

```python
%%sql

CREATE VIEW view_q4_revenue_earned_not_earned_loss AS
SELECT round(sum(total_price_usd), 2) as total_revenue, 
(SELECT round(sum(total_price_usd), 2)
FROM view_revenue_analysis
WHERE paid_status = 1 and is_delivered = 1) AS revenue_earned,
(SELECT round(sum(total_price_usd), 2)
FROM view_revenue_analysis
WHERE paid_status = 1 and is_delivered = 0) AS revenue_not_earned,
(SELECT round(sum(total_price_usd), 2)
FROM view_revenue_analysis
WHERE paid_status = 0 and is_delivered = 0) AS revenue_order_placed_unpaid_undelivered,
(SELECT round(sum(total_price_usd), 2)
FROM view_revenue_analysis
WHERE paid_status = 0 and is_delivered = 1) AS loss_delivered_not_paid
FROM view_revenue_analysis
LIMIT 10;
```

```python
%%sql

SELECT *
FROM view_q4_revenue_earned_not_earned_loss
```

Q5. Taking into consideration that some orders were not completed, how much money was ACTUALLY generated compared to the answer from Question 1?

```python
%%sql

CREATE VIEW view_q5_revenue_actually_generated AS
SELECT product, count(*) as units_ordered, 
(SELECT ROUND(SUM(total_price_usd), 2)
FROM view_revenue_analysis
WHERE paid_status = 1
AND product = r.product) as revenue_received,                     
ROUND(SUM(total_price_usd), 2) as potential_revenue,
round((SELECT ROUND(SUM(total_price_usd), 2)
FROM view_revenue_analysis
WHERE paid_status = 1
AND product = r.product)/ROUND(SUM(total_price_usd), 2)*100) as percentage_of_revenue_actually_generated
FROM view_revenue_analysis r
GROUP BY product;
```

```python
%%sql

SELECT *
FROM view_q5_revenue_actually_generated
ORDER BY percentage_of_revenue_actually_generated DESC;
```
